```python
img_1 = cv2.imread('5.png')
```

```python
imgbox = pytesseract.image_to_boxes(img_1) #location of the text in the image
print(imgbox)
```

```python
imgH,imgW,_ = img_1.shape
```

```python
img_1.shape
```

```python
for boxes in imgbox.splitlines():
    boxes = boxes.split(' ')
    x,y,w,h = int(boxes[1]),int(boxes[2]),int(boxes[3]),int(boxes[4])
    cv2.rectangle(img_1,(x,imgH-y),(w,imgH-h),(0,0,225),3) 
```
